>Using the data below, answer the following questions:

>1) Which entities (top 5) had the largest population density in 2020?
>2) Which entities have more water area than land area?
>3) Which entities increased in population the most in the last 10 years?

>6) Which city has the largest percent drop from their largest city to their 5th largest? 100*(1st largest - 5th largest)/(1st largest)

```python
import pandas as pd
```

```python
facts = pd.read_csv('../data/state_facts.tsv',delimiter="\t")
#facts.head(5)
```

```python
#1. 
facts['Popdensity']=facts['Pop_2020']/facts['Area_tot']
print(facts.sort_values(by='Popdensity',ascending=False).head(5))
#District of Columbia, New Jersey, Rhode Island, Massachusetts, Connecticut
```

```python
#2.
facts[(facts['Area_water']>facts['Area_land'])]
```

```python
#3.
facts['increase'] = facts['Pop_2020']-facts['Pop_2010']
print(facts.sort_values(by='increase',ascending=False).head())
#Texas, Florida, California, Georgia, Washington
```

```python
#4.
facts.sort_values(by='Pop_2020',ascending=False).head()
#California Quail, Northern Mockingbird, Northern Mockingbird, Eastern Bluebird, Ruffed Grouse
```

```python
#4. 
facts.sort_values(by='Area_land', ascending=False).head()
#Alaska, Texas, California, Montana, New Mexico
```

```python
facts.columns
```

```python
#5.
facts[facts['Capital']==facts['City_1']]
#17
```

```python
#6. 
facts['Large_percentdrop']= 100*(facts['city_1_pop']-facts['city_5_pop'])/facts['city_1_pop']
facts.sort_values(by='Large_percentdrop',ascending=False).head()
#New York
```
