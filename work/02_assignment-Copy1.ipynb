```python
import pandas as pd
```

```python
facts = pd.read_csv('../data/state_facts.tsv',delimiter="\t")
#facts.head(5)
```

```python
#1. 
facts['Popdensity']=facts['Pop_2020']/facts['Area_tot']
print(facts.sort_values(by='Popdensity',ascending=False).head(5))
#District of Columbia, New Jersey, Rhode Island, Massachusetts, Connecticut
```

```python
#2.
facts[(facts['Area_water']>facts['Area_land'])]
```

```python
#3.
facts['increase'] = facts['Pop_2020']-facts['Pop_2010']
print(facts.sort_values(by='increase',ascending=False).head())
#Texas, Florida, California, Georgia, Washington
```

```python
#4.
facts.sort_values(by='Pop_2020',ascending=False).head()
#California Quail, Northern Mockingbird, Northern Mockingbird, Eastern Bluebird, Ruffed Grouse
```

```python
#4. 
facts.sort_values(by='Area_land', ascending=False).head()
#Alaska, Texas, California, Montana, New Mexico
```
