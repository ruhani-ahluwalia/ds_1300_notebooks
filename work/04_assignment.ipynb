```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Country.csv',delimiter=',',skipfooter=14,engine='python')
```

```python
df
```

```python
df = df[df['date'].str.contains('-21', na=False)]
df
```

```python
months=[]
for i in range (len(df)):
    month= str(i)+' month'
    months.append (i)
```

```python
df = df[df['date'].str.contains('-21', na=False)]
```

```python
print (months)
df['number_of_months'] = months
```

```python
df
```

```python
df.plot.scatter(x = 'number_of_months', y = 'cases', color= 'red', figsize=(20,2)) 
plt.title("COVID-19 Cases per Month from 2020-2023")
plt.xlabel("Number of Months")
plt.ylabel("Cases")
plt.legend(["Confirmed cases"], loc ="upper left")
```
